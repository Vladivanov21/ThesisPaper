# Building A Neural Network Based Classifier FOr Predicting Drug Response From Single-Cell Expression Profiles Of Tumours

## Building A Neural Network Model

First imports and loading the data:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.linear_model import LinearRegression

final_expression = pd.read_csv('processed_data/final_expression.csv')
drug_sensitivity = pd.read_csv('processed_data/drug_sensitivity.csv')
MKN_data = pd.read_csv('processed_data/MKN_data.csv')
NCIH_data = pd.read_csv('processed_data/NCIH_data.csv')
SCC_data = pd.read_csv('processed_data/SCC_data.csv')
SKMEL_data = pd.read_csv('processed_data/SKMEL_data.csv')
```

```python
X = final_expression.drop(columns=['cell_line']).values  # Features (gene expressions)
y = drug_sensitivity['(+)-CAMPTOTHECIN (GDSC2:1003) 0.1μM rep1'].values # We are using only one drug for simplicity

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.05, random_state=42
)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# 3. Define PyTorch model class
class RegressionModel(nn.Module):
    def __init__(self, input_dim, hidden1, hidden2, activation):
        super(RegressionModel, self).__init__()
        act_fn = nn.ReLU() if activation == 'relu' else nn.Tanh()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden1),
            act_fn,
            nn.Linear(hidden1, hidden2),
            act_fn,
            nn.Linear(hidden2, 1)
        )

    def forward(self, x):
        return self.net(x)

# 4. Training function
def train_model(model, dataloader, optimizer, loss_fn, epochs):
    model.train()
    for epoch in range(epochs):
        for X_batch, y_batch in dataloader:
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = loss_fn(y_pred, y_batch)
            if torch.isnan(loss):
                print("Loss is NaN — stopping training")
                break
            loss.backward()
            optimizer.step()

# 5. Evaluation function
def evaluate_model(model, X_test, y_test):
    model.eval()
    with torch.no_grad():
        y_pred = model(X_test)
        mse = mean_squared_error(y_test.numpy(), y_pred.numpy())
        r2 = r2_score(y_test.numpy(), y_pred.numpy())

        y_pred_np = y_pred.numpy()
        y_true = y_test_tensor.numpy()

    reg = LinearRegression().fit(y_true, y_pred)
    y_fit = reg.predict(y_true)

    plt.scatter(y_true, y_pred_np, alpha=0.6)
    plt.plot(y_true, y_fit, color='green', label='Best-Fit Line')
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', label='Perfect Line')
    plt.xlabel("True Drug Sensitivity")
    plt.ylabel("Predicted")
    plt.title("True vs. Predicted Drug Sensitivity")
    plt.grid(True)
    plt.show()

    return mse, r2

# 6. Grid Search manually with ParameterGrid
param_grid = {
    'hidden1': [32, 64],
    'hidden2': [16, 32],
    'activation': ['relu', 'tanh'],
    'lr': [0.001, 0.01],
    'batch_size': [16, 32],
    'epochs': [90, 100]
}

best_score = float('inf')
best_params = None
best_model = None

for params in ParameterGrid(param_grid):
    print(f"Training with params: {params}")
    model = RegressionModel(X_train.shape[1], params['hidden1'], params['hidden2'], params['activation'])
    optimizer = optim.Adam(model.parameters(), lr=params['lr'])
    loss_fn = nn.MSELoss()
    
    dataset = TensorDataset(X_train_tensor, y_train_tensor)
    dataloader = DataLoader(dataset, batch_size=params['batch_size'], shuffle=True)
    
    losses = train_model(model, dataloader, optimizer, loss_fn, epochs=params['epochs'])
    mse, r2 = evaluate_model(model, X_test_tensor, y_test_tensor)
    
    print(f"--> MSE: {mse:.2f}, R²: {r2:.2f}")
    if mse < best_score:
        best_score = mse
        best_params = params
        best_model = model
        best_losses = losses

print("\n✅ Best Params:", best_params)
final_mse, final_r2 = evaluate_model(best_model, X_test_tensor, y_test_tensor)
print(f"Final Test MSE: {final_mse:.2f}, R² Score: {final_r2:.2f}")
```

```python
variance = y.var()
print(f"Variance of the target variable: {variance:.2f}")
```

We have very little variance of our target variable, so it makes snese to have 0 R^2 score, so our model performs well with MSE 0.07.

We save the best model and we are good to go to the evaluation.

```python
torch.save(best_model.state_dict(), 'best_model.pth')
```
